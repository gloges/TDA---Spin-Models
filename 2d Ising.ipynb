## Computing persistence images
PIs are constructed from persistence diagram data, which has been saved/loaded as (birth, persistence, frequency). With a lattice many points in the persistence diagram occur with precisely the same birth/death, so saving their frequency helps with efficiency.

A weight log(1+p) is used to kill low-persistence features/emphasize long-lived features which are more "real". Having the weight vanish for p=0 is part of the stability theorems.

```python
# pdlocnumI gives index of persistence diagram in loaded data
# bounds should be a list [bmin,bmax,pmin,pmax]
# bins should be a list [bbins, pbins]
# stdev gives width of gaussians

def PerIm(pdlocnumI,bounds,bins,stdev,res=1000):
    if pdlocnumI % 5000 == 0:
        print("%6.2f %%" % (100 * pdlocnumI / len(PD1data)))
    
    pdlocnum = PD1data[pdlocnumI]
    
    x = np.linspace(bounds[0], bounds[1], res)
    y = np.linspace(bounds[2], bounds[3], res)
    xx, yy = np.meshgrid(x, y, sparse=True)
    z=np.zeros([res,res])
    
    for i in np.arange(len(pdlocnum)):
        birth = pdlocnum[i][0]
        persistence = pdlocnum[i][1]
        weight = np.log(1 + persistence)
        freq = pdlocnum[i][2]
        z += freq * weight * np.exp((-0.5*(xx-birth)**2-0.5*(yy-persistence)**2)*stdev**-2)
    
    return z.reshape(bins[0], res//bins[0], bins[1], res//bins[1]).sum(3).sum(1)
```

```python
# Create persistence images
bounds = [0, 10, 0, 10]
bins = [20, 20]
res = bins[0] * 10
sigma = 0.5

imgs1 = [PerIm(p, bounds, bins, sigma, res) for p in np.arange(len(PD1data))]
print("100.00 %\n\nDONE")
```

Flatten persistence images into 1D arrays and normalize according to the L1 norm (so they are probability distributions).

```python
# Flatten into 1D vectors
imgs1_array = np.array([img.flatten() for img in imgs1])

# Normalize
imgs1_array = preprocessing.normalize(imgs1_array, norm='l1')
```

## Birth-Death distributions, correlation length and critical exponents
Compute 1D statistics in birth-death plane by projecting onto a line eminating from the origin.
The distributions do not change much as the angle changes. Death distributions (90 deg) is perhaps best motivated.

```python
# project x, a list of [b,p,f] onto a line in b,d(=p+b) space
def projection(x, angle):
    r = (x[:,0]**2 + (x[:,1]+x[:,0])**2)**0.5
    theta = np.arctan(x[:,1]/x[:,0] + 1)
    return r*np.cos(theta - angle)
```

```python
# Create distributions for each temp
valueBinWidth = 1
maxValue = 55

# Angle in birth-death space
#   angle =  0   <->   birth
#   angle = 90   <->   death
angle = np.deg2rad(90)

# value -> center of bin
value = np.arange(0, maxValue, valueBinWidth) + valueBinWidth / 2
valueFreq = np.zeros([len(tempsUnique), len(value)])
valueTotals = np.zeros(len(tempsUnique))

for i in range(len(tempsUnique)):
    indices = np.where(TPD1list == tempsUnique[i])[0]

    # birth, death, frequency
    bdf = PD1data[indices]
    
    # Clean up odd outliers (those with b=d)
    bdf = [x[np.where(x[:,1] > 0)[0]] for x in bdf]

    # value, frequency
    vf = [np.transpose([projection(x, angle), x[:,2]]) for x in bdf]
    
    for run in vf:
        for pt in run:
            valueBin = min(int(pt[0]//valueBinWidth), len(value)-1)
            count = int(pt[1])
            valueFreq[i, valueBin] += count
    
    # normalize
    valueTotals[i] = sum(valueFreq[i])
    valueFreq[i] /= sum(valueFreq[i])
```

```python
# Plot some of the distributions
tempsToPlot = [1.8, 2.1, 2.4, 2.7, 3.0]
indices = np.where([(t == tempsToPlot).any() for t in tempsUnique])[0]


fig, ax = plt.subplots(1,1, figsize=(8,4))
for i in indices:
    # shift bin (central) values by binWidth/2 for the purposes of plotting
    ax.step(value - valueBinWidth/2, valueFreq[i], where='post')

plt.yscale('log')
plt.xlim([0, maxValue])

plt.xlabel('death')
plt.ylabel('death frequency')
# plt.xlabel('value')
# plt.ylabel('value frequency')
plt.legend([("T = %.2f" % t) for t in tempsToPlot])
plt.show()
```

### Fit distributions to exponential

```python
# log(exp(A)*x^(-mu)*exp(-k*x))
def expLaw(x, mu, k):
    return -k*(x-0.5) - mu * np.log(x/0.5)
```

```python
muList = np.zeros(len(valueFreq))
muErrList = np.zeros(len(valueFreq))
invXiList = np.zeros(len(valueFreq))
invXiErrList = np.zeros(len(valueFreq))

for i in range(25, len(valueFreq)):
    yData = valueFreq[i]
    nonzeroCounts = np.where(yData > 10**-5)[0][0:]

    xData = value[nonzeroCounts]
    yData = yData[nonzeroCounts]

    popt, pcov = curve_fit(expLaw, xData, np.log(yData))

    muList[i] = popt[0]
    muErrList[i] = np.sqrt(pcov[0,0])
    invXiList[i] = popt[1]
    invXiErrList[i] = np.sqrt(pcov[1,1])

    
n = 25
xFit = np.linspace(0.1, value[-1], 100)
yFit = np.exp(expLaw(xFit, muList[n], invXiList[n]))

plt.plot(xFit, yFit)

plt.step(value - valueBinWidth/2, valueFreq[n], where='post')
plt.yscale('log')
plt.ylim([10**-6, 2])
plt.show()

inds = range(25, len(tempsUnique))

fig, ax = plt.subplots(1,2, figsize=(12,4), sharex=True)
ax[0].errorbar(tempsUnique[inds], muList[inds], muErrList[inds], ls='', marker='o', ms=5, c='k', elinewidth=2, capsize=3)
ax[1].errorbar(tempsUnique[inds], invXiList[inds], invXiErrList[inds], ls='', marker='o', ms=5, c='k', elinewidth=2, capsize=3)

ax[0].axvline(x=2.37, c='r', ls='--', lw=1)

ax[0].set_xlabel('T')
ax[1].set_xlabel('T')
ax[0].set_ylabel(r'$\mu$')
ax[1].set_ylabel(r'$1/\xi$')

plt.show()
```

## Betti numbers
Not wildly different from death distribution in this case...

```python
maxValue = 50
numBins = 50

binLeftEdges = np.linspace(0, maxValue, numBins)
bettiCounts = np.zeros([len(tempsUnique), numBins])


for j in range(len(tempsUnique)):
    t = tempsUnique[j]

    indices = np.where(TPD1list == t)[0]

    for i in indices:
        bdf = PD1data[i]

        for pt in bdf:
            birth = pt[0]
            persistence = pt[1]
            freq = pt[2]

            minIndex = int((birth / maxValue) * numBins)
            maxIndex = int(((birth + persistence) / maxValue) * numBins)

            if maxIndex > len(bettiCounts[j]):
                maxIndex = len(bettiCounts[j])

            bettiCounts[j][minIndex:maxIndex] += freq

    bettiCounts[j] /= len(indices)
```
